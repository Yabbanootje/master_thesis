```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(10)
baseline_data = np.random.rand(10) * 100  # Random data for baseline
comparison_data = baseline_data + np.random.randn(10) * 10  # Baseline with some random deviation

# Calculate the deviations
deviations = comparison_data - baseline_data

# Plot the baseline line
plt.plot(x, baseline_data, label='Baseline', color='black', linestyle='--')

# Plot the deviation bars
plt.bar(x, deviations, bottom=baseline_data, label='Deviation', color='skyblue')

# Add labels and legend
plt.xlabel('Instance')
plt.ylabel('Value')
plt.title('Deviation Bar Chart')
plt.legend()

# Show plot
plt.show()
```

```python
# Sample data
x = np.arange(10)
baseline_value = 100  # Constant baseline value
comparison_data = baseline_value + np.random.randn(10) * 10  # Baseline with some random deviation

# Calculate the relative deviations (percentage difference)
relative_deviations = (comparison_data - baseline_value) / baseline_value * 100

# Plot the baseline line
plt.axhline(y=0, color='black', linestyle='--', label='Baseline')

# Plot the deviation bars
plt.bar(x, relative_deviations, label='Deviation', color='skyblue')

# Add labels and legend
plt.xlabel('Instance')
plt.ylabel('Relative Deviation (%)')
plt.title('Deviation Bar Chart')
plt.legend()

# Show plot
plt.show()
```
